```python
import numpy as np
import matplotlib.pyplot as plt
#to see things inline
%matplotlib inline
```

```python
#to create a 2d grid
points = np.arange(-512, 512)
x_c, y_c = np.meshgrid(points, points)
```

```python
#x_c is x coordinates of grid, y_c is y coordinates of grid
print x_c, y_c
```

```python
z = np.sin(x_c) + np.cos(y_c)
```

```python
#to show colorscale
plt.imshow(z)
plt.colorbar()
plt.imshow(z)
plt.title('sine + cos')
```

```python
#zip function explanation zip combines two lists and iterates over it zip([a, b], [c,d]) is [(a,c), (b,d)]
a = [1, 2, 3, 4]; b= [3, 4, 5, 6]; c = [True, False, True, False];
z = [a_val if c_val == True else b_val for a_val, b_val, c_val in zip(a, b, c)]
print z
```

```python
#where is similar to ?: in sv and C
z = np.where(c, a, b);
print z
```
